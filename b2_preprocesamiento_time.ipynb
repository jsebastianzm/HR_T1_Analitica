```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import a_funciones as fn
import datetime
# import 
```

```python
in_time = 'data\in_time.csv'
out_time = 'data\out_time.csv'

#Se procesa los dataframe con pandas y se almacenan
d_time_in = pd.read_csv(in_time)
d_time_out = pd.read_csv(out_time)
```

```python
d_time_in.head()
d_time_in.rename(columns={'Unnamed: 0':'EmployeeID'}, inplace=True)
d_time_in.head()
```

```python
# d_time_out.head()
d_time_out.rename(columns={'Unnamed: 0':'EmployeeID'}, inplace=True)
d_time_out.head()
```

```python
d_time_in.isnull().sum()
```

```python
d_time_out.isnull().sum()
```

```python
for i in d_time_out.columns[1:]:
    d_time_in[i] = pd.to_datetime(d_time_in[i], format='%Y-%m-%d %H:%M:%S')
    d_time_out[i] = pd.to_datetime(d_time_out[i], format='%Y-%m-%d %H:%M:%S')
```

```python
d_time_full = []
d_time_full = pd.DataFrame(d_time_full)
for i in d_time_out.columns[1:]:
    duracion = d_time_out[i].dt.hour - d_time_in[i].dt.hour
    d_time_full[i] = duracion
```

```python
d_time_full.head()
```

```python
d_time_mean = pd.DataFrame()
# d_time_full.mean(axis=1)
d_time_mean["EmployeeID"] = d_time_in["EmployeeID"]
d_time_mean["mean_time"] = d_time_full.mean(axis=1)
```

```python
d_time_mean.head()
```

```python
d_time_mean['mean_time'] = d_time_mean['mean_time'].round(0)
```

```python
d_time_mean['mean_time'] = d_time_mean['mean_time'].astype(int)
```
